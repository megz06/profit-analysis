```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r"C:\Users\ELCOT\OneDrive\Desktop\my_fst_project\archive\Sample_ Superstore.csv",encoding='latin1')
print("Profit by Category")
category_profit=df.groupby("Category")["Profit"].sum()
print(category_profit)
print("\nProfit by Region")
region_profit=df.groupby("Region")["Profit"].sum()
print(region_profit)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r"C:\Users\ELCOT\OneDrive\Desktop\my_fst_project\archive\Sample_ Superstore.csv",encoding='latin1')
fig,(ax1,ax2)=plt.subplots(1,2, figsize=(12,5))
category_profit.plot(kind="bar",ax=ax1, color=["blue","green","orange"])
ax1.set_title("Profit by Category")
region_profit.plot(kind="bar",ax=ax2, color=["cyan","pink","red","purple"])
ax2.set_title("Profit by Region")
plt.tight_layout()
plt.savefig("Profit_analysis.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r"C:\Users\ELCOT\OneDrive\Desktop\my_fst_project\archive\Sample_ Superstore.csv",encoding='latin1')
subcategory_profit=df.groupby("Sub-Category")["Profit"].sum().sort_values(ascending=False).head(5)
print("Top 5 Most Profitable Sub-Categories")
print(subcategory_profit)
subcategory_profit.plot(kind="bar",color="blue")
plt.title("Top 5 Profitable Sub_Categories")
plt.xlabel("Sub Category")
plt.ylabel("Profit")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("subcategory_profit.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r"C:\Users\ELCOT\OneDrive\Desktop\my_fst_project\archive\Sample_ Superstore.csv",encoding='latin1')
sns.boxplot(x="Category",y="Profit",data=df)
plt.title("Profit Distribution by category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.tight_layout()
plt.savefig("boxplotprofit_analysis.png")
plt.show()
```
